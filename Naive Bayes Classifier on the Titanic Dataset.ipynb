# **Naive Bayes Classifier on the Titanic Dataset**

```python
# loaded the Titanic dataset to a dataframe
import pandas as pd

input = pd.read_csv('/kaggle/input/titanic-machine-learning-from-disaster/train.csv')


```

```python
input.head()
```

```python
#removing unnecessary columns

input.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
input.head()
```

```python
input_y=input["Survived"]
input_x=input.drop("Survived",axis='columns')

input_x.head()
```

```python
dummies = pd.get_dummies(input_x["Sex"])

dummies.head()
```

```python
input_x = pd.concat([input_x,dummies],axis='columns')

input_x.head()
```

```python
input_x.drop("Sex",axis='columns',inplace=True)
input_x.head()
```

```python
# Check if any null values are there
input_x.isnull().sum()
# train_x.isnull().any()
```

```python
# Listing the null rows
input_x[input_x.isnull().any(axis='columns')] 
```

```python
#replace null values with mean

input_x.Age = input_x.Age.fillna(input_x.Age.mean())

input_x.head()
```

```python
input_x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(input_x,input_y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB

model =GaussianNB()
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
pred = model.predict(test_x)
```
